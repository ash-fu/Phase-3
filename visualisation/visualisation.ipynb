# Elements of Visualization
All forms of visualization are built with some basic visual elements such as:

- Location (x,y coordinate in the screen)
- Brightness
- Color (Hue)
- Pattern/Texture
- Shape
- Line
- Text

Visualization, in principle, transforms the numerical and symbolic data into these basic visual elements. In the previous example, the cities are translated into the colors of the lines and the temperature data are used to plot the location of the lines.

From those simple elements, some popular types of visualization can be built such as:

- graphs, representation of a set of relationships between various entities, like family trees, network diagrams, grammar trees
- maps, representation of a particular space (and its properties), like geographical map, brain activity map
- charts, representation of numerical data either from a given set of real-world data or generated by mathematical functions

In this worksheet, you will mostly learn to generate various types of charts: line plot (line chart), bar chart, pie chart, histogram, and box plot.

# The Structure of matplotlib

The `matplotlib` is conceptually divided into three parts. The first part, matplotlib API, is the library that does the hard-work, managing and creating figures, text, lines, plots and so on. In the code above, we access this library by issuing the following command:

    >>> import matplotlib

The device dependent backend is the second part. It is the drawing engine that is responsible to render the visual representation to a file or a display device. Example of backends: 'PS' backend is used to create postscript file (suitable for hardcopy printout), 'SVG' creates scalar vector graphics (SVG file), 'Tkinter' on Windows provides interactive interface to the visualization. In IVLE, you will use Agg backend to produce PNG file, as displayed the example above:

     >>> matplotlib.use('Agg')

The `pyplot` interface is the last part of the matplotlib package. Module `pyplot` provides a set of functions that utilize the underlying matplotlib library. High level visualization functions like plot, boxplot, and bar, are available through pylab interface. To import these functions, issue the following command:

    >>> import matplotlib.pyplot as plt

## The Powerful Plot
Plot is probably the most important function in matplotlib. Plot draws lines and/or markers using coordinates from the multiple points or x,y pairs supplied in the argument of the function. Both x and y are generally list or array of values. For example, the following command plot a simple quadratic function `y = x * x`.

    >>> x = [1,2,3,4]
    >>> y = [1,4,9,16]
    >>> plot(x,y)

A single list argument to the plot command, like `plot(y)`, would be considered as a list of y-values. matplotlib will automatically generates the x-values for you. This is displayed in the next example, which plots the monthly averages of maximum temperature in Melbourne (See table in the beginning of this worksheet).

```python
%matplotlib inline
import matplotlib.pyplot as plt
import calendar

# Melbourne maximum temperature Mar 2007 - Feb 2008	
t = [41.2,35.5,37.4,29.3,23.9,16.8,18.2,25.7,22.3,33.5,36.9,41.1]
# print 'jan' - 'dec'
plt.xticks(range(12),list(calendar.month_abbr)[1:])
plt.plot(t) 
```

## Plot line properties

You can supply an optional argument to customize the color and the linestyle of plot output. For example, to plot with red circles, you would issue `plot([1,2,3,4],'ro')`. `'r'` represents red color and `'o'` refers to circle-shaped marker. `plot([1,2,3,4],'bs:')` draws blue dotted line with square marker. The line color, the linestyle, and the marker type are respectively given by `'b'`,`':'`, and `'s'`. Select your preferred color from a set of matplotlib colors. The choices for linestyle and marker can be seen in the table below.

**Line properties**

Property | Values
- | -

antialiased | True or False - use antialised rendering
color | a matplotlib color arg
label | a string optionally used for legend
linestyle | One of -- : -. -
linewidth | a float, the line width in points
marker | One of + , o . s v x > < ^
markeredgewidth | The line width around the marker symbol
markeredgecolor | The edge color if a marker is used

There are other methods to set plot line properties. First, you can use keyword arguments listed in the table above. For example:

    >>> plot(x, y, linewidth=3.0)

Second method uses `setp()` command. Shown below, `setp()` allows you to modify multiple properties of a collection of lines.

    >>> lines = plot(x1, y1, x2, y2)
    >>> setp(lines, color='b', linewidth=4.0)

`plot()` may return a list of lines; eg line1,line2 = plot(x1,y1,x2,x2). Third method utilizes various set functions on the lines returned by a plot command. The list of the set functions are available [here](http://matplotlib.org/api/lines_api.html).

    >>> line1,line2 = plot(x1, y1, x2, y2)
    >>> line1.set_antialiased(False) # turn off antialising on the first line

The use of those three methods are demonstrated in the following example:

```python
%matplotlib inline
import matplotlib.pyplot as plt
from numpy import arange

t = arange(0.0, 2.05, 0.05)

# method 1: keyword arguments
plt.plot(t, sin(t*10), 'k-', linewidth=3.0)
# plot a solid black line with thickness=3

# method 2: setp command
lines = plt.plot(t, [4 for i in t], t, 4*t)
plt.setp(lines, 'color', 'r' )
plt.setp(lines, 'linestyle', ':' )
# plot two red dotted lines

# method 3: setp command
line1,line2 = plt.plot(t, t**2, t, exp(t))
line1.set_marker('s')
line1.set_color('g')
line2.set_marker('^')
line2.set_color('b')
# plot two lines
# first line is drawn with green square marker
# second line is drawn with blue triangle marker
```

# Scatter plot

With real world data, `plot()` function can be used to generate time series and scatter plot.

Scatter plot is often used to display the relationship between two variables (plot as x-y pairs). In this scatter plot example, we use [famous Iris data set](http://en.wikipedia.org/wiki/Iris_flower_data_set). The data is available [here](data/iris.csv). This data set provides measurements on various parts of three types of Iris flower (Iris setosa, Iris versicolour, and Iris virginica). For each type, there are 50 measurements, or samples. Each data row in the CSV file contains (1) sepal length, (2) sepal width, (3) petal width, (4) petal length, and (5) the type of Iris flower.

The following code generates the scatter plot between petal length and petal width of the three Iris types.

```python
import matplotlib.pyplot as plt
from numpy import arange
import urllib
import csv

source = open("data/iris.csv","r")
data = list(csv.reader(source))
source.close()
# use dictionary for keeping the list
# petal_length['Iris-setosa'] contains a list of measurements
# petal_length['Iris-setosa'] = [4.0 4.0 3.0 etc] 
petal_length = {} 
petal_width = {}

for row in data:
    # 3 species: Iris-setosa, Iris-versicolour, or Iris-virginica
    species = row[4]
    if species not in petal_length:
        petal_length[species] = []
        petal_width[species] = []	
    # third column contains petal length
    petal_length[species].append( float(row[2]) ) 
    # fourth column contains petal width
    petal_width[species].append( float(row[3]) ) 

for species in petal_length: # for all 3 species
    # call plot function to generate scatter plot
    # use "o" argument to generate circle marker
    plt.plot( petal_length[species], petal_width[species], "o" )

all_species = petal_length
plt.legend(all_species,loc='lower right', numpoints=1)    
plt.ylabel("petal width")
plt.xlabel("petal length")
```

From the scatter plot, we may be able to suggest a particular type of relationship or a formation of clusters. In the example above you may notice that, for Iris versicolor, the samples with longer petal tend to have wider petal. You can also see clearly that there exist clusters of these three Irises. As such, the measurements of petal and sepal can help identifying the type of Iris flower. This example demonstrates how botanists may indentify a certain species from phenotype characteristics.

# Time series

Time series is used to display historical data over a period of time. From time series, we can often deduce trend, predict the future data, and observe relationship between different time periods. The following example plots monthly historical data of Australia interest rate, set by Reserve Bank of Australia every month.

```python
%matplotlib inline
import matplotlib.pyplot as plt
from numpy import arange
import csv

# data from Jan 1990 to Mar 2008
source = open("data/au_interest_rates.csv","r")
data = list(csv.reader(source))
source.close()

hist_rates = [] # historical rates 
years = [] # ticks text for year
y_loc = [] # ticks location
prev_idx = 0 # index to previous month

for arow in data:
    rate = float(arow[1]) # interest rate
    date = arow[0].split('/') # data of interest rate change
    month = int(date[0])-1 # map jan-dec to 0-11
    year = int(date[2])-1990 # map 1990-2008 to 0-18 
    idx = year*12 + month
    for i in arange(prev_idx, idx+1): # filling earlier months
        hist_rates.append( rate ) # build hist_rates
        if i%12 == 0: # if january build years ticks
            y_loc.append( i )
            years.append( str(1990+year) )
    prev_idx = idx+1

plt.xticks(y_loc,years,rotation='vertical' )
plt.plot(hist_rates)
```

Check [the CSV file](data/au_interest_rates.csv) and you will notice that the data is sparse, not all months from Jan-1990 to March-2008 have interest rate entry. So, you need to build a complete array of monthly rates throughout that period. The nested loop in the given code just does that by producing `hist_rates` list.

## Horizontal bar chart

Horizontal bar chart is basically normal bar chart rotated 90 degrees. It is useful to display properties of a particular set with relatively larger number of entities. The example below shows the distribution of characters in the first paragraph of Moby Dick:

```python
%matplotlib inline
import matplotlib.pyplot as plt

text = open('data/moby.txt').read()
tally = zeros(26) # initialize frequencies of 26 characters
charset = [chr(i) for i in arange(65,65+26)] # 'A'-'Z'
for char in text:
    ch = char.upper()
    if ch in charset:
        tally[ord(ch)-65] += 1

plt.barh(arange(26),tally, color=(0.3,0.4,0.4), linewidth=0)
plt.yticks(arange(26),charset)
plt.ylim(0,26)
```

*Note:* In the given code, we use `ord()` function. This function converts a character into its numerical representation (ASCII code), for example ord('A')=65, ord('B')=66, ord('C')=67, and so on.

## Plot date

`matplotlib` provides a special function called [`plot_date()`](http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.plot_date) that can be used to plot a list of values associated with a list of corresponding dates (similar to time series). In the time series example above, you have to manually fill in the gaps between the dates when interest rates change. With `plot_date()` function, you don't have to do that; you need not to worry about the placement and the labelling of the date axis. The function just asks you to supply two arguments: (1) the list of dates and (2) the list of corresponding values to be plot.

The following example shows you how to use plot_date() to plot daily closing values of ASX 300. The data is available as a [CSV-formatted file](data/asx300.csv). On top of that, the example also plots a set of markers to display stock buys on some particular dates. The list of dates are kept in another text file.

```python
%matplotlib inline
import matplotlib.pyplot as plt
from datetime import date
import calendar, csv, sys

# creates a list of 'Jan'-'Dec'
threeLettersMonth = list(calendar.month_abbr)

# two data files
buyDatesFile = "data/buy_dates.txt"
asx300File = "data/asx300.csv"

# function to produce a list from a CSV file
def getList(filename):
    f = open(filename)
    data = list(csv.reader(f))
    f.close()
    return data

# function that reads file buy_dates.txt -------------------------
# format:
# 29 Jan 2008
# 23 Jan 2008

def readBuyDates():
    dateList = []    
    buyDateTable = getList(buyDatesFile)
    buyDateTable.reverse() # sort by date ascending
  
    for aRow in buyDateTable:
        fields = aRow[0].strip().split(" ") # get date from text 
        year = int(fields[2])
        month = threeLettersMonth.index(fields[1])
        day = int(fields[0])
        aBuyDate = date( year, month, day )
        dateList.append( aBuyDate )

    numDateList = [date2num(i) for i in dateList]
    return numDateList

# function that reads file asx300.csv ----------------------------
# format:
# Date,Open,High,Low,Close,Volume,Adj Close
# 22/02/2008,5589.6,5589.6,5498.7,5567.4,0,5567.4
# 21/02/2008,5504.6,5607.3,5504.6,5589.6,0,5589.6

def readAsxData():
    dateList = []
    closeList = []    
    asxDataTable = getList(asx300File)[1:]
    asxDataTable.reverse() # sort by date ascending
    
    for aRow in asxDataTable:
        closeValue = float(aRow[6])
        fields = aRow[0].strip().split("/") # get date from text 
        year = int(fields[2])
        month = int(fields[1])
        day = int(fields[0])
        aBuyDate = date( year, month, day )
        dateList.append( aBuyDate )    
        closeList.append( closeValue )

    numDateList = [date2num(i) for i in dateList]
    return (numDateList,closeList)
    
# main program
# 1. plot ASX 300 index from 2006-2008 --------------------------
(asxDateList,asxCloseList) = readAsxData()
plot_date(asxDateList, asxCloseList, 'b-')

# 2. plot only specific buy dates --------------------------------
buyDates = readBuyDates()
buyDateList = []
buyPriceList = []

for aBuyDate in buyDates: # for each buy date, find its price
    if aBuyDate in asxDateList:
        vIndex = asxDateList.index(aBuyDate)
        buyDateList.append( aBuyDate )
        buyPriceList.append(asxCloseList[vIndex])

plt.plot_date(buyDateList, buyPriceList, 'rs', markersize=10)
# customize the output
plt.xticks(rotation=-30, ha='left' )
plt.grid(True)
```

## Box plot

Box plot is used to show the detail distribution of measured variables. The example below displays the boxplot of two measurements: the petal width of Iris setosa and the petal width of Iris virginica (See scatter plot section for introduction of this Iris stuff).

```python
%matplotlib inline
import matplotlib.pyplot as plt
import csv

data = csv.reader(open('data/iris.csv'))
petal_length = {}
petal_width = {}

for row in data:
    # Iris-setosa, Iris-versicolour, or Iris-virginica
    species = row[4]
    if species not in petal_length:
        petal_length[species] = []
        petal_width[species] = []	
    # third column contains petal length
    petal_length[species].append( float(row[2]) ) 
    # fourth column contains petal width
    petal_width[species].append( float(row[3]) ) 

# boxplot two variables:
# Iris-setosa's petal_width and Iris-virginica's petal_width
plt.boxplot( [petal_width['Iris-setosa'], petal_width['Iris-virginica']] )
plt.ylim(0,3)
plt.xticks( arange(3), ['','Iris-setosa','Iris-virginica'] )
plt.ylabel( "petal width" )
```

You can notice that that Iris virginica has longer box than Iris setosa, which means there is more variation of petal width in Iris virginica samples. You can confirm this by examining the example of scatter plot (The cluster of Iris virginica occupies a larger area than one of Iris setosa)

## Figure and Subplot

All the examples so far used implicit figure and axes creation. Figure is kind of container for your plot, and axes is the rectangle area that bounds the x-axis and y-axis.

This basic plot command:

    >>> plot([1,2,3])

is equivalent to:

    >>> figure() # first figure (implicit)
    >>> subplot(111) # only 1 plot in the figure (implicit)
    >>> plot([1,2,3])
    
You can use the functions `figure()`, `subplot()`, and `axes()` to explicitly retrieve the figures and axes, and have finer grain control over them. For example, you can change the size of the matplotlib result using a command `figure(figsize=(width,height))` or you can change the color of the plot area using: `subplot(111,axisbg=(0.6,0.7,0.8) )`.

`subplot()` allows you to create multiple axes in a figure. Subplot subdivides a figure into rows and columns of axes. For example, to create two rows and one column of axes, you would issue `subplot(211)` to create the upper axes and `subplot(212)` for the lower axes. The first and second digit indicates the number of the rows and the column respectively. The last digit indicates the index of the axes counted across the rows. The following example uses the Iris data set to generate 2 by 2 scatter plot.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import csv

data = csv.reader(open('data/iris.csv'))
petal_length = {}
petal_width = {}
sepal_length = {}
sepal_width = {}

for row in data:
    # Iris-setosa, Iris-versicolour, or Iris-virginica
    species = row[4]
    if species not in petal_length:
        petal_length[species] = []
        petal_width[species] = []
        sepal_length[species] = []
        sepal_width[species] = []
    # first column contains sepal length
    sepal_length[species].append( float(row[0]) ) 
    # second column contains sepal width
    sepal_width[species].append( float(row[1]) ) 
    # third column contains petal length
    petal_length[species].append( float(row[2]) ) 
    # fourth column contains petal width
    petal_width[species].append( float(row[3]) ) 

plt.subplot(221)
for species in petal_length:
    plt.plot( petal_length[species], petal_width[species], "o" )
plt.ylabel("petal width")

plt.subplot(222)
for species in petal_length:
    plt.plot( sepal_length[species], petal_width[species], "o" )

plt.subplot(223)
for species in petal_length:
    plt.plot( petal_length[species], sepal_width[species], "o" )
plt.xlabel("petal length")
plt.ylabel("sepal width")

plt.subplot(224)
for species in petal_length:
    plt.plot( sepal_length[species], sepal_width[species], "o" )
plt.xlabel("sepal length")

all_species = petal_length
plt.legend(all_species,loc='lower right', numpoints=1)
```
